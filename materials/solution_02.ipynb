### 1. Consider the system $$x_{n+1} = \cos(x_n),$$ where $x_0\in\mathbb{R}$

### 1.1. How many fixed points does this system has?

```python
def f(x):
    return np.cos(x)

x = np.linspace(0, 1.5, num=100)
plt.plot(x, x, color='black')
plt.plot(x, f(x), color='black')
plt.grid()
plt.show()
```

### 1.2. Find all fixed points of this system.

```python
def f(x):
    return np.cos(x)

x = np.linspace(0, 1.5, num=100)
n = 10
x = [None] * n
x[0] = 0
for k in range(n-1):
    x[k+1] = f(x[k])
    
fp = x[-1]
print(f'Fixed-point: fp = {fp}') 
```

### 1.3. Show cobweb plots

```python
def f(x):
    return np.cos(x)

x = np.linspace(0, 1.5, num=100)
plt.plot(x, x, color='black')
plt.plot(x, f(x), color='black')

n = 10
x = [None] * n
# Y = [None] * n
x[0] = 0
plt.plot(x[0], x[0] ,marker='x', markersize=10, color='blue', label='Seed')
for k in range(n-1):
    x[k+1] = f(x[k])
    px = x[k]
    py = x[k+1]
    plt.plot([px, px], [px, py], color='red')
    plt.plot([px, py], [py, py], color='red')
    plt.plot(px, py, marker='o', color='red')
    plt.plot(px, px, marker='x', color='blue')
    
plt.legend()
plt.grid()
plt.show()
```

### 2. Consider the system $$x_{n+1} = 0.02x_n^2-12.5,$$ where $x_0\in\mathbb{R}$

### 2.1. How many fixed points does this system has?

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2-(x/6)+(1/3)

x = np.linspace(0, 1, num=100)
plt.plot(x, x, color='black')
plt.plot(x, f(x), color='black')
plt.grid()
plt.show()
```

### 2.2. Find all fixed point of this system and the cobweb plots

```python
def f(x):
    return x**2-(x/6)+(1/3)

x = np.linspace(0, 0.6, num=100)
plt.plot(x, x, color='black')
plt.plot(x, f(x), color='black')

n = 100
x = [None] * n
# Y = [None] * n
x[0] = 0.1
plt.plot(x[0], x[0] ,marker='x', color='red', label='Seed')

for k in range(n-1):
    x[k+1] = f(x[k])
    xx = x[k]
    yy = x[k+1]
    plt.plot([xx, xx], [xx, yy], color='red')
    plt.plot([xx, yy], [yy, yy], color='red')
    plt.plot(xx, yy, marker='o', color='red')

plt.legend()
plt.grid()
plt.show()

fp1 = x[-1]
print(f'Fixed-point: fp1 = {fp1}') 
```

```python
def g(x):
    return np.sqrt((7*x/6)-(1/3))

x = np.linspace(0.5, 1, num=100)
plt.plot(x, x, color='black')
plt.plot(x, g(x), color='black')

n = 100
x = [None] * n

x[0] = 1
plt.plot(x[0], x[0] ,marker='x', color='red', label='Seed')

for k in range(n-1):
    x[k+1] = g(x[k])
    xx = x[k]
    yy = x[k+1]
    plt.plot([xx, xx], [xx, yy], color='red')
    plt.plot([xx, yy], [yy, yy], color='red')
    plt.plot(xx, yy, marker='o', color='red')

plt.legend()
plt.grid()
plt.show()

fp2 = x[-1]
print(f'Fixed-point: fp2 = {fp2} --> f({fp2}) = {f(fp2)}') 
```

```python
# Mann's Iteration

def f(x):
    return x**2-(x/6)+(1/3)

alpha = lambda n: 1/2
seed = 0
n = 1000
x = [None] * n
x[0] = seed
for k in range(n-1):
    x[k+1] = (1-alpha(k))*x[k] + alpha(k)*f(x[k])
print(x[-1], f(x[-1]))
```

### 5. Consider the system

$$
\begin{align}
x_{n+1} &= \frac{1}{y_n}(10 - x_n^2), \\
y_{n+1} &= 57-3x_ny_n^2,
\end{align}
$$ 

where $(x_0,y_0)\in\mathbb{R}^2$.


### 5.1. How many fixed points does this system has?

```python
import numpy as np
import matplotlib.pyplot as plt


def f1(x,y):
    return (1/y)*(10-x**2)

def f2(x,y):
    return 57-3*x*y**2

    
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
plt.contour(X, Y, f1(X, Y)-X, [0])
plt.contour(X, Y, f2(X, Y)-Y, [0])
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

### 5.2. Find all fixed points of the system

```python
def f1(x,y):
    return np.sqrt(10-x*y)

def f2(x,y):
    return np.sqrt((57-y)/(3*x))


def T(x,y):
    return np.array([f1(x,y), f2(x,y)])

n = 100
z = np.zeros(shape=(n, 2))
seed = np.array([1, 1])
z[0,:] = seed
for k in range(n-1):
    
    xnow, ynow = z[k,0], z[k,1]

    xnew = f1(xnow, ynow)
    xnow = xnew
    
    ynew = f2(xnow, ynow)
    ynow = ynew
    
    z[k+1,:] = np.array([xnow, ynow])

fp = z[-1,:]
print(fp, '--->', T(*fp))
```

### 6. Consider the system
$$
\begin{align}
x_{n+1} & = \frac{1}{5}\left( \sin(xy) -x^2 - y \right), \\
y_{n+1} & = \frac{1}{6}\left( \cos(x+y) + y^2 -x +2 \right),
\end{align}
$$ 

where $(x_0,y_0)\in\mathbb{R}^2$.

### 6.1. How many fixed points does this system has?

```python
import numpy as np
import matplotlib.pyplot as plt


def f1(x,y):
    return (1/5)*(np.sin(x*y)-x**2-y)

def f2(x,y):
    return (1/6)*(np.cos(x+y)+y**2-x+2)

    
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)
plt.contour(X, Y, f1(X, Y)-X, [0])
plt.contour(X, Y, f2(X, Y)-Y, [0])
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

### 6.2. Find all fixed points of this system.

```python
def f1(x,y):
    return (1/5)*(np.sin(x*y)-x**2-y)

def f2(x,y):
    return (1/6)*(np.cos(x+y)+y**2-x+2)

def T(x,y):
    return np.array([f1(x,y), f2(x,y)])

alpha = lambda n: 3/4

seed = np.array([1, 1])

n = 10
z = np.zeros((n, 2))
z[0:] = seed
for k in range(n-1):
    xnow, ynow = z[k,0], z[k,1]
    
    xnew = (1-alpha(k))*xnow + alpha(k)*f1(xnow, ynow)
    xnow = xnew
    
    ynew = (1-alpha(k))*ynow + alpha(k)*f2(xnow, ynow)
    ynow = ynew
    
    z[k+1,:] = np.array([xnow, ynow])

fp = z[-1,:]
print(fp, '--->', T(*fp))
```
